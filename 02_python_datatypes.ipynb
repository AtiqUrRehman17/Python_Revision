## Python Data Types
Here In python there are three types
1. `Basic Type` (Integer,Float,String,Bolean,And Complex)
2. `Cantainer Type` (List,Tuple,Set,Dictionary)
3. `User Defined Type `- Class

## **`Basic Type`** : Basic Types are like we know that there are many types in python like Integer , Float , String etc

## **`Container Type`** : here in container type we store myltiple data types like int , Float , string etc

-----

# Basic Type Example

```python
# 1 integer 
# python can handle lots of interger like the range of integer is 1e308
print(10)
print(1e308)
```

```python
# 2 Float 
# Float is also same as integer but in decimal format
print(3.3)
# also the same range for Float as well
print(1.7e308)
```

```python
# 3 Bolean 
print(True)
print(False)
```

```python
# 4 Complex 
# here this is the advatage of python cos in many language there is no support of the complex dtype
print(4+6j)
print(1+3j)
```

```python
# 5 String
print('Pakistan')
print("Hello World")
print('''Pakistan''')
```
