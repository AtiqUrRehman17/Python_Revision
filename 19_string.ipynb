### isalnum/isalpha/isdecimal/isdigit/isidentifier
these all are like questional function where it used to check whatever it is 

```python
# isalpha it used to check whatever the string all elements are alpha numaric or not
'FLAT20'.isalnum()
```

```python
# alpha it is used to check that all the elements in the string is alphabetical or not
'FALT'.isalpha()
```

## Split Function :

```python
# here the split function is used to split the string and put it into a list like]
'who is the pm of pakistan'.split()
# here it split the string and put it into a list
```

```python
# here u can also split the string on the basis of the user given like
'who is the pm of pakistan'.split('pm')
# here it split on the basis of pm and comes two strings
# but the condition is that the elements shulod be avliable in the string
```

```python
# JOIN Function
# basically it is the reverse of the split function 
# exaple is that suppose u have list and u want make it again a string then u can use the join function
```

```python
# here how to join the list by the Sapce
' '.join(['who is the pm of pakistan'])
```

```python
# here u can also join it by anything like
'-'.join(['who', 'is', 'the',  'pm', 'of', 'pakistan'])
```
