## Fraction Datatype
We Will Create our own fraction datatype

```python
class Faraction:

    def __init__(self, n, d):
        self.num = n
        self.deno = d

    def __str__(self):
        return '{}/{}'.format(self.num, self.deno)
    
    # now lets use another magic method for adding

    def __add__(self, other):
        temp_num = self.num * other.deno + other.num * self.deno
        temp_deno = self.deno * other.deno 
        return '{}/{}'.format(temp_num,temp_deno)
    
    def __sub__(self, other):
        temp_num = self.num * other.deno - other.num * self.deno
        temp_deno = self.deno * other.deno 
        return '{}/{}'.format(temp_num,temp_deno)
    
    def __mul__(self, other):
        temp_num = self.num * other.num
        temp_deno = self.deno * other.deno 
        return '{}/{}'.format(temp_num,temp_deno)
    
    def __truediv__(self, other):
        temp_num = self.num * other.deno 
        temp_deno = self.deno * other.num
        return '{}/{}'.format(temp_num,temp_deno)


```

```python
f1 = Faraction(1, 2)
f2 = Faraction(3, 4)

print(f1 + f2)
print(f1 - f2)
print(f1 * f2)
print(f1 / f2)
```

## Encapsulation

```python
class Atm:
    # constructor
    def __init__(self):
        self.__pin = ''
        self.__balance = 0

        self.menu()

    def get_pin(self):
            print(self.__pin)

    def set_pin(self,new):
            if type(new) == str:
                self.__pin = new
                print('pin Chnaged')
            else:
                print('Not Allowed')


    def menu(self):
        user_input = input('''
                How would you like to proceed
                1.enter 1 to create pin
                2.enter 2 to deposite
                3.enter 3 to withdraw
                4.enter 4 to check balance
                5.enter 5 to exit
                ''')

        if user_input == '1':
            self.create_pin()
        elif user_input == '2':
            self.deposite()
        elif user_input == '3':
            self.withdraw()
        elif user_input == '4':
            self.check_balance()
        else:
            print('bye')

    # now we have to write methods for each functionality

    def create_pin(self):
        user_pin = int(input('enter your pin'))
        self.__pin = user_pin
        print('pin created successfully')
        self.menu()

    def deposite(self):
        temp = int(input('enter your pin please'))
        if temp == self.__pin:
            amount = int(input('enter amount please'))
            self.__balance = self.__balance + amount
            print('deposite successfull')
        else:
            print('invalid pin')
        self.menu()

    def withdraw(self):
        temp = int(input('enter your pin please'))
        if temp == self.__pin:
            amount = int(input('enter amount please'))
            if amount < self.__balance:
                self.__balance = self.__balance - amount
            else:
                print('insufficent Funds')
        else:
            print('invalid pin')
        self.menu()

    def check_balance(self):
        temp = int(input('enter your pin please'))
        if temp == self.__pin:
            print(self.__balance)
        else:
            print('invalid pin')
```
