### Heigher Order Function

```python
# here we will create a list and we will make a function which will give us three output
l = [11,14,22,23,78,29,28,33]
# 1 is that the even numbers in the string we want the sum of that
# 2 same the odd numbers
# 3 is that the numbers which is divisable also sum of that

# here the code that will be
def return_sum(l):
    even_sum = 0
    odd_sum = 0
    div3 = 0
    for i in l:
        if i %2 == 0:
                even_sum += i
                if i %2 != 0:
                    odd_sum += i
                if i % 3 == 0:
                    div3 += i
    return(even_sum,odd_sum,div3)
```

#### Now for the above code we can improve it by the HOF method

```python
# for even
x = lambda x:x%2==0
# for odd
y = lambda x : x % 2 != 0
div3 = lambda x : x%3 ==0
# where l is
l = [11,14,22,23,78,29,28,33]
# Here we konw that HOf means whcih used a function in input
def return_sum(func,l):
    
    result = 0
    
    for i in l:
        if func(i):
            result = result + i
    return(result)

print(return_sum(x,l))
print(return_sum(y,l))
print(return_sum(div3,l))

```

### 1. Map
Map is a function which take two inputs one is `function (lambda)` and 2 is `itrable` (list,tuple,string etc)
Map is used if u give a itrable (like List) so in each element u can perform operations.

```python
# exaple is we have a list and we want to dubble the elements of the list
l = [1,2,3,4,5,6]
# so this can done easily by the map function like
print(l)
print(map(lambda x:x**2,l)) # the ouput is differnt so we have to do type convertion
print(list(map(lambda x:x**2,l)))
```

```python
# lets do anather exaple of like we check that the elements are odd or even
list(map(lambda x:'even' if x%2 == 0 else 'Odd',l))
```

### Filter
here in map we can perform opration on the itrabel each element but the Filter filter the data

```python
# here suppose we have a list and we want to print those elements which are greater then 4
l = [1,2,3,4,5,6,7]
list(filter(lambda x:x>4,l))
# alomst same but on the elements we can't perform operation just we can filter it out
```

```python
# lets say we have anther list of frutes and we want to print only those froutes in whcih e is occur
froutes = ['Apple','Orange','Mango','Gauva']
list(filter(lambda x: 'e' in x,froutes))
```

### Reduce
