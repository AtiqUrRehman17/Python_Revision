```python
def anagram(str1,str2):
    str1 = str1.replace(' ','')
    str2 = str2.replace(' ', '')
    
    return sorted(str1) == sorted(str2)
```

```python
anagram('hello','olleh')
```

```python
anagram('atiq','qita')
```

```python
anagram('khan','atiq')
```

```python
def multi_lis(num):
    product = 1
    for i in num:
        product *= i
    return product
```
