```python
def is_anagram(str1, str2):
    # Convert both strings to lowercase and remove spaces
    str1 = str1.replace(" ", "").lower()
    str2 = str2.replace(" ", "").lower()
    
    # Compare sorted characters of both strings
    return sorted(str1) == sorted(str2)
```

```python
a = 'silent'
b = 'listen'
print(is_anagram(a,b))
```

```python
def max_min(number):
    min = number[0]
    max = number[0]
    
    for num in number:
        if num < min:
            min = num
            if num > max:
                max = num
    return min,max
    
```

```python
print(max_min([1,2,3,4,5,6]))
```

```python
def my_task(arr):
    max_length = len(arr[0])
    for item in arr:
        if len(item) > max_length:
            max_length = len(item)

    longest_strings = []
    for item in arr:
        if len(item) == max_length :
            longest_strings.append(item)
        return longest_strings

```

```python
a = ['Pakistan','india','Turkey','Germany','America']
print(my_task(a))
```

```python
b = ["Dubai","india","Turkey","Germany","America"]
print(my_task(b))
```
