## Dynamic Typing
In the other languages Like C Java Etc , We declare the Variable In The Beggning Where in Python We Dont Need 
To Define or Declare the Variable In the Beggning , So this is called `Dynamic Typing`

## Static Typing :
Static Typing is the oppsite of Dynamic typing in which like in other Languages we declare the Variable In the Beggning.
So this is called `Static Typing`

```python
# here in python the variable that is defined so if you give the same variable a differnt vlue so the varible will update
name = 4
name
# Here if u see the name variable is updated and store the Latest value
```

```python
# same here 
name = True
name
```

### Speacial Declaration Syntax Used In Python

```python
# here if you want to store myltiple variables in same line so u can used the follow syntax Just use the ; in the middle
a = 5; b = 'atiq'; c = False
print(a)
print(b)
print(c)
```
