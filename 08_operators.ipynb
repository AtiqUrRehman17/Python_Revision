1. `Arithematic Operator` (-,+,*,/) etc

```python
x = 5
y = 2
print(x + y)  # addition
print(x - y) # subtraction 
print(x * y)  # multiplication
print(x / y) # divition
print(x % y) # modulus the remainder 
print(x ** y) # power of the opetrator
print(x // y) # integer divition used to convert the coeffient in to inger if it is in float
```

### `Comparision Operator ` (<,>,=,==,!=,<=,>=) etc

```python
print(x > y)
print(x < y)
print(x <= y)
print(x >= y)
print(x == y)
print(x != y)
```

### `Logical Operator` (& , |,  !) (end , or , not )

```python
x = True
y = False
#  Or Operator
print( x or y)
# it cames true when one on the operation is ture
```

```python
# And Operator
print(x and y)
# it comes only True when both operations are True
```

```python
# Not operator
print(not y)
# opposite result will give by this operation
```

### `Bitwise Operators` ( It only Works on Binary Values)

```python
x = 2
y = 3
print(x & y)
print(x >> 2)
print(x << 3)
# here we use the bitwise & not the Logicaly where we canvert the x and y values in to binary
```

### `Identity Operator` (Basically it checked that is ther is two variables on the same memory location or not)

```python
a = 3
b = 3
print(a is b)
# its true it means that both variables are in the same memory location
```

```python
a = 4 
b = 5
print(a is b)
# False it means that both variables are not in the same location
```

```python
a = 'hello'
b = 'hello'
print(a is b)
```

```python
a = 'helo'
b = 'hello'
print(a is b)
```
