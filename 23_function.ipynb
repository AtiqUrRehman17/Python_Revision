```python
def is_even(number):
    if number % 2==0:
        return 'Even'
    else:
        return 'Odd'
```

## Lets See How to Create Arbitrary Function

```python
# lets create a function which will give the product of the number if one input the one output if the number is
# more then it will multilpy the numbers and give us a result
def flexi(*number):
    product = 1
    for i in number:
        product = product * i
    print(product)
```

### Nested Function
Nested Function means That you are creating anather function iside the function

```python
def f():
    print('inside f')
    def g():
        print('Inside g')
    g()
f()
```
