# Sets In Python :
Sets are the same that we studied in metric Like the uniion and Intersection
- **`Reules For Sets`** :
- Sets do not allow duplicates
- Sets Have No Indexing / Slicing
- Sets Do not allow mutable data types
- Sets itself is a Mutable datatype

```python
# Here Lets see how to create Sets
# here creating a empty set is basically not a set basically its a dicnationary
s1 = {}
print(type(s1))
```

```python
# here now u can create a empty set like
s2 = set()
print(type(s2))
```

```python
# Homogenous set
s1 = {1,2,3,4,5}
print(s1)
```

```python
# hetrogenous set
s2 = {'hello',1,1.1,True}
# so here the indexing is change which means set do no allow indexing so it can be on the basis of set
print(s2)
```

```python
# here we will see the first Rule which is that set donot allow duplicates
s3 = {1,1,2,2,3,3}
print(s3)
# so here automatically the duplication of elemnets are removed.
```

```python
# here now we will see the third rule which is sets donot allow mutable datatype
s3 = {[1,2,3],'hello'}
print(s3)
# so here it givem an error becasea list is mutable datatype
```

```python
# now let see the same rule for tuple which is immutable data type
s4 = {(1,2,3),'hello'}
print(s4)
# now here it is printed becase tuple is immutable datatype
```

```python
# here lets see the 2d sets whcih are not possiable in sets
s5 ={{1,2},{3,4}}
print(s5)
```

```python
# here lets see the indexing and slicing which is not possiable in sets
s2 = {1,2,3,4,5}
s2[0]
# here it means thta indexing or nagative indexing and slicing are not working in sets.
```

### Here editing is not possiable in sets so what if we convert a set in list by type convertion and make the editin and reconvert the list in set so is it worked no becase the memory location is changed.

```python
# let take a set and convert that into a list
s1 = {1,2,3,4}
# lets look its memory location where it stored
id(s1)
```

```python
# now lets convert the set to a list and make changes
s1 = list(s1)
s1
```

```python
# now lets add 100 instead of 1
s1[0]=100
print(s1)
```

```python
# now lets reconvert the list into set
s1 = set(s1)
print(s1)
# also check its memory location which is changed
print(id(s1))
```

```python
# now lets talk about addition can we add new item the existing set the answer is yes
s1 = {1,2,3,4,5}
print(s1)
# lets add 6 to the set
s1.add(6)
print(s1)
```

```python
# here u can also remove elements from a set
# delete
# remove
# pop
```

```python
# deleting eniter set 
del(s1)
```

```python
# here u can also delete elemnts from a set by the del function
del s2[0]
# by the del function it will apply on the whole set but we cannot delete a single element from a set due to indexing
```

```python
# here by the reomve function u can remove any element from a set
s1 = {1,2,3,4,5}
print(s1)
s1.remove(5)
print(s1)
```

```python
# here u can use the pop operation which will delete the last elemnt fom the set
s1.pop()
print(s1)
```

```python
# SET OPERATIONS :
# here the firest operation is concatenation on sets u cannot perform concatenations like;
s1 = {1,2,3,4,5}
s2 = {5,6,7,8,9}
s1 + s2
```

```python
# Also you canot perform multiplication on setss
s1 * s2
```

```python
# But u can perform the loops on sets
for i in s1:
    print(i)
```

```python
# similar membership operators ar also worked in set
1 in s1
```

```python
# FUNCTIONS : basically the basic functions are also work on sets
print(len(s1))
print(min(s1))
print(max(s1))
print(sum(s1))
print(sorted(s1,reverse=True))
```

### Here there are some spacific function that is used only by sets :

```python
print(s1)
print(s2)
# here we will see the union and intersection of the sets
print(s1.union(s2))
print(s1.intersection(s2))
```

```python
# here there is one more function which is called difference which is used to check the element which is different from 
# each other Like :
s1.difference(s2)
```

```python
s2.difference(s1)
```

```python
# here symmatric_differnce which is used to check those elements which are not present in boths
s1.symmetric_difference(s2)
```

```python
# here isdijoint function which is used to check the common elements between the sets
s1.isdisjoint(s2)
```

```python
# issubset whcih mean one set have all the elemnts of the other set
s1.issubset(s2)
```

```python
# issuperset which is used to check that the first set is the superset of the other set
s1.issuperset(s2)
```
