```python
# Lets See how to create List :
# 1. Empty List
l =[]
print(l)
# 2. Homogenous List (same dtype):
l = [1,2,3,4,5]
print(l)
# 3. Hetrogeous List (not same dtype):
l = ['atiq',True,1,22.2]
print(l)
```

```python
# Multi Dimentioal List it means List within a list :
# 2D List :
L = [1,2,3,[4,5]]
print(L)
# 3D List :
L = [1,2,[4,6],[6,7]]
print(L)
```

```python
# Here We Will See That how To Acces Item From List :
# here accessing the item form a list is same to the string everything is same to string 
l = [1,2,3,4,5]
print(l[0])
# Or
print(l[2])
# reverse accessing 
print(l[-1])
print(l[2:4])
print(l[::-1])
```

```python
# here the little difference from string is the 2D and 3D list 
# like extract the element from 2d list
L = [1,2,3,[4,5]]
# now here how to extract the element
L[-1]
# here the Last element index is -1 and it comes the second list in the list
# so lets store the second the list in the list with a variable
x = L[-1]
print(x[0])
# here we can also access the string like this
print(L[-1][0])
```

```python
# Exaple of 3D List :
L4 = [[[1,2],[3,4]],[[5,6],[7,8]]]
print(L4)
# now here how to acces the the 9 from the 3D Lsit here the same Square Brakets u will use
print(L4[1][1][0])
# here i want to print 3
print(L4[0][1][0])
```

```python
# Here we will see that how to add New items to list
# Here we can add new item to list by three way
# 1. Append
# 2.Extend
# 3.insert
```

```python
# 1.Append append is used toa dd new item in the last of the list
a = [1,2,3,4]
print(a)
# here i want to add 5 to the list
a.append(5)
print(a)
# here the 5 is added to the list in the list
```

```python
# Extend it is used to add multiple thing in the last of the string
# like i want to add hello and 6 to the list
a.extend(['hello','atiq',6])
print(a)
```

```python
# Insert is used to something to list in a desired index means to that index that u want
b = [1,2,3,4,5]
# here i want to add hello world between 1 and 2 so
b.insert(1,'hello world')
# here in the inser function u will give two thing one is the index u want to come the new thing and second is what u 
# want to intsert like the above
print(b)
```

```python
# here we will see that how to delete things from a list
# here we can delete something from a list like the words that is del,remove,pop,clear here we will use each one
# here if u want to delete the entire list then
del(b) # b list is deleted
```

```python
# here u can also delete a portion from a list
del b[5] # 5 index of hello world is deleted
print(b) 
```

```python
# here this can be done by reverse indexing
c = [1,2,3,4,5,6]
del c[-2] # 5 will be deleted
print(c)
```

```python
# now remove it is used when u dont konw the index but u know that i want to delete then reove is used
c.remove(6)
print(c)
```

```python
# last one is pop it is used to delete the last index number
c.pop() # 4 will be deleted
print(c)
```

```python
# clear basically it clear the list means it reomve everything from the list
c = [1,2,34,5]
print(c)
print(c.clear())
```
