##### 1 Range Function :

##### so this is built in function in python which is used to generate integers numbers in given numbers
##### range(1,10)
##### here the above range is generated as it is but internally 1 to 10 numbers is generated if u store it into

```python
range(1,10)
```

```python
list(range(1,10))
```

```python
# here there are basically three values in the for loop
# where the first value is start value the second value is the end value and teh third is step
# but if u give only one value to the range function then it means its the end value and it take the first
# value 0
list(range(5))
```

```python
# here the third value in the range function which is step it is used for the difference between the numbers
# it depends on the user whatever he increase or decrease the value
list(range(1,10,2))
# here it means print values from 1 to 10 and the difference is 2 between the numbers 
```

```python
# here u can also used the function for backword counting
list(range(10,0,-1))
```

### Sequence :
1. Here in Python there is one more concept of sequence  and sequence means order
2. here if you get something order so it is called sequence like  string 'pakistan'
3. so basically it is a sequence of charctor .
4. sequence can string ,tuple ,set,dictionary,etc

```python
# here why we understnad these two things Becase the is either itrate on the range or sequence
for i in range(1,11):
    print(i)
# it actually itrate on the range function
```

```python
# same the step function
for i in range(1,11,2):
    print(i)
```

```python
# here lets see the for loop on sequence
for i in 'Pakistan':
    print(i)
# here how it works
# here the first value of i is the first charctor of the string and it itrate untill the string end
```
