# **Tuples In Python**
Here tuples are much similar to List but There is some mojor different that we will see

```python
# here we will that how to create tuples 
# 1. Empty Tuples.
t = ()
print(t)
```

```python
# Homogenous Tuple mean which have sama data type elements
t1 = (1,2,3,4,5)
print(t1)
```

```python
# hetrogenous Tuple mean which have different data typle elements
t2 = ('hello',1,1.1,False)
print(t2)
```

```python
# same like list u can create 2d and 3d typles
t4 = (1,2,3,(2,3))
t4
```

```python
# here the difference is that if u assing only one element to the tuple then it will not be a tuple
t5 = ('hello')
t6 = (1)
print(type(t5))
print(type(t6))
# here it not showing that its a tuple.
```

```python
# so whenever u create a tuple with a single element the alway remeber that u will assina comma after the element
t7 = ('hello',)
print(t7)
print(type(t7))
```

```python
# u can also do the type convertion like
t8 = tuple('hello')
print(t8)
print(type(t8))
```

```python
# here u can also convert a list to a tuple
t9 = tuple([1,2,3,4,5])
print(t9)
```

```python
# here we will see that how to access the a tuple 
# here every thing is exctly same to list till here
t9[0]
print(t9[1:4])
print(t9[-1])
print(t9[::-1])
```

### Here the difference is that when you edit the list so it will be edited but tuple can't.

```python
# suppose we have list and i want to change its first element
l = [1,2,3,4]
print(l)
# here i wnat to replace the 1 with 100
l[0]=100
print(l)
```

#### Here it mean that Tuples are Like string u cannot edit it once its maked then u can't edit it.Same also that u cannot add new items to tuples.

```python
# Now deletion tuples here u can delete the tuple by the del function
del t1
# here the t1 will be deleted
```

```python
# here u cannot delete a single element from the tuple because if one elemnt is deleted then in the tuple changes will ocur
t11 = (1,2,3,4,5)
del t11[-1]
```

```python
# Operation On Tuple :
# here operation are same like List 
# 1. Concatenation
# 2.multiplication
# 3.Loops can be performed 
# 4. Also Membership opetrators can be performed
t4 +t11
print(t4 * 3)
for i in  t11:
    print(i)
1 in t11
```

```python
# function are also same
print(len(t11))
print(min(t11))
print(max(t11))
```

```python
# also sorted and sort can be performed
sorted(t11,reverse=True)
```

### So here the mojor difference between tupe and list is that List is reand and write data type which means thatyou can access and make changes But Tuple is Read-only data type which means you can only access the tuple But u can't make changes.
