### Rules Of Dictionary
1. Dictionary has no indexing
2. Dictionary is a Mutable Data type
3. Keys shuold always be immutable Where Values can be mutable
4. Keys Should be unique

#### Mutable Data type Vs Immutable Datatype
Mutable is > List / Dictionary / Sets
Immutable is > String / Tuple / Int / Float / etc

```python
# here lets see how to create a dic
d = {}
print(d)
d1 = {'Name':'atiq','age':21,'From':'Bannu'}
print(d1)
```

```python
# lets ee the 3d rule which is that keys will be immutable
d2 = {[1,2,3]:'name'}
print(d2)
# error and same hashing concept is used
```
